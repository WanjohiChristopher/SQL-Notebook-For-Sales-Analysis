 #### This notebook is a simple monthly sales analysis using SQL from scratch. The notebook uses various SQL statements, to perform different data science data processing analysis.

```python
import pandas as pd
import numpy as np
```

```python
import pymysql
#Establishing db connection with notebook
conn=pymysql.connect(host='localhost',port=int(3306),user='root',passwd='',db='officesupplies')
```

```python
#Let's retrive the records in our table
sales = pd.read_sql("SELECT * FROM sales_january_2019;", conn)
sales.head(100)
```

```python
from datetime import datetime
sales['Order_Date']=pd.to_datetime(sales['Order_Date'])

```

```python
#Checking the duplicates
sales = pd.read_sql('SELECT Product,Quantity_Ordered  FROM sales_january_2019 GROUP BY Product,Quantity_Ordered HAVING count(*) > 1;', conn)
sales
```

```python
#Now we can delete the duplicates in dataset
sales = ('DELETE FROM sales_january_2019 WHERE field in (SELECT Product,Quantity_Ordered FROM sales_january_2019  GROUP BY Product,Quantity_Ordered HAVING count(*) > 1);', conn)
sales
```

```python
#Let's group our products regionally 
regions = pd.read_sql("SELECT * FROM sales_january_2019 ORDER BY Purchase_Address ;", conn)
regions
```

```python
#Let's print the iphone customers only in the rregion
EastClients = pd.read_sql("SELECT Purchase_Address,Product FROM sales_january_2019 WHERE Product = 'iPhone';", conn)
EastClients
```

```python
#Let's print our 377 Meadow St, New York City, NY 10001 customers alone
CentralClients = pd.read_sql("SELECT Order_Date, Purchase_Address,Product FROM sales_january_2019 WHERE Purchase_Address = '377 Meadow St, New York City, NY 10001';", conn)
CentralClients
```

```python
#Print only the 2019 january clients
#https://www.plus2net.com/sql_tutorial/between-date.php
clients_2019 = pd.read_sql('SELECT Order_Date FROM sales_january_2019 WHERE Order_Date >= '2019-01-13 16:43:00' and Order_Date <= '2019-01-22 21:25:00');', conn)
clients_2019   
```

```python
#How many total products count ? 
totalpricesales= pd.read_sql('SELECT Product,count(*) as count FROM sales_january_2019 GROUP BY Product order by 2 DESC;', conn)
totalpricesales
```

```python
#let's select product sold above 100 only
priceabove100=pd.read_sql('SELECT Product as count FROM sales_january_2019 WHERE Price_Each >= 100.00;',conn)
priceabove100
```

```python
#How many total products were soled in the year 2019 january alone?
sales_2019= pd.read_sql('SELECT Order_Date FROM sales_january_2019 WHERE Order_Date BETWEEN '2019-01-13 16:43:00 AND 2019-01-22 21:25:00';',conn)
sales_2019
```

```python
#How many iphone were soled ?
totaliphones=pd.read_sql("SELECT COUNT(Product) FROM sales_january_2019 WHERE Product='iphone';", conn)
totaliphones
```
